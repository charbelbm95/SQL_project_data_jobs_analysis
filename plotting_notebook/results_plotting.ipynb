```python

import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
from collections import Counter
```

```python

query_1_results = pd.read_csv('../results/1_top_paying_jobs.csv')
query_2_results = pd.read_csv('../results/2_top_paying_jobs_skills.csv')
```

### Query 1

```python
query_1_results.head()
```

```python
plt.figure(figsize=(10, 6))
plt.barh(query_1_results.job_title, query_1_results.salary_year_avg, color="skyblue")
plt.xlabel("Average Annual Salary ($)")
plt.ylabel("Job Title")
plt.title("Top 10 Remote Data Analyst Jobs by Salary")
plt.gca().invert_yaxis()  
plt.tight_layout()
plt.show()
```

### Query 2

```python
skills_counts = Counter(query_2_results['skills'])
skills_names = list(skills_counts.keys())
skills_frequencies = list(skills_counts.values())
```

```python
plt.figure(figsize=(12, 8))
plt.barh(skills_names, skills_frequencies, color="lightblue")
plt.xlabel("Frequency")
plt.ylabel("Skills")
plt.title("Frequency of Skills in Top-Paying Data Analyst Jobs")
plt.tight_layout()
plt.gca().invert_yaxis()  # Invert y-axis for better readability
plt.show()
```
